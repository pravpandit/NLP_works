## 1. पर्यावरण सेटिंग

पिछले अध्याय की तरह ही, हमें सबसे पहले OpenAI API का उपयोग करके पर्यावरण को कॉन्फ़िगर करना होगा

```python
import openai
#तृतीय-पक्ष लाइब्रेरी आयात करें

openai.api_key  = "sk-..."
# API_KEY सेट करें, कृपया इसे अपनी API_KEY से बदलें

# निम्नलिखित पर्यावरण चर पर आधारित कॉन्फ़िगरेशन विधि का एक उदाहरण है, जो अधिक सुरक्षित है। यह केवल संदर्भ के लिए है और बाद में इसे कवर नहीं किया जाएगा।
#आयात openai
# ओएस आयात करें
# OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY")
# openai.api_key = OPENAI_API_KEY
```

```python
def get_completion_from_messages(messages, 
                                 model="gpt-3.5-turbo", 
                                 temperature=0, 
                                 max_tokens=500):
    '''
    封装一个访问 OpenAI GPT3.5 的函数

    参数: 
    messages: 这是一个消息列表，每个消息都是一个字典，包含 role(角色）和 content(内容)。角色可以是'system'、'user' 或 'assistant’，内容是角色的消息。
    model: 调用的模型，默认为 gpt-3.5-turbo(ChatGPT)，有内测资格的用户可以选择 gpt-4
    temperature: 这决定模型输出的随机程度，默认为0，表示输出将非常确定。增加温度会使输出更随机。
    max_tokens: 这决定模型输出的最大的 token 数。
    '''
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=temperature, # 这决定模型输出的随机程度
        max_tokens=max_tokens, # 这决定模型输出的最大的 token 数
    )
    return response.choices[0].message["content"]
```

### 2.1 संबंधित उत्पाद और श्रेणी के नाम निकालें

अपने ग्राहक की क्वेरी को वर्गीकृत करने के बाद, आपको क्वेरी की श्रेणी मिल जाएगी - चाहे वह खाता समस्या हो या उत्पाद समस्या हो। फिर आप विभिन्न श्रेणियों के आधार पर अलग-अलग कार्रवाई कर सकते हैं।

प्रत्येक उपकार्य में केवल संबंधित कार्य को करने के लिए आवश्यक निर्देश होते हैं, जो सिस्टम को प्रबंधित करना आसान बनाता है, यह सुनिश्चित करता है कि मॉडल में कार्य करने के लिए आवश्यक सभी जानकारी है, और त्रुटियों की संभावना कम हो जाती है। यह दृष्टिकोण लागत को भी कम कर सकता है, क्योंकि लंबे समय तक संकेत और अधिक टोकन के परिणामस्वरूप चलने की लागत अधिक होती है, और कुछ मामलों में सभी चरणों की रूपरेखा तैयार करना आवश्यक नहीं हो सकता है।

इस दृष्टिकोण का एक अन्य लाभ यह है कि इससे यह परीक्षण करना आसान हो जाता है कि कौन से चरण विफल होने की अधिक संभावना हो सकती है, या किसी विशिष्ट चरण में मानवीय हस्तक्षेप की आवश्यकता हो सकती है।

जैसे-जैसे आप इन मॉडलों का निर्माण और इनके साथ इंटरैक्ट करते हैं, आपमें यह अंतर्ज्ञान विकसित होगा कि इस रणनीति का उपयोग कब करना है। इसके अलावा, एक अतिरिक्त लाभ के रूप में, यह मॉडल को आवश्यकता पड़ने पर बाहरी उपकरणों का उपयोग करने की अनुमति देता है। उदाहरण के लिए, यह उत्पाद सूची में कुछ खोजने, एपीआई को कॉल करने, या ज्ञान आधार खोजने का निर्णय ले सकता है, जो एकल प्रॉम्प्ट का उपयोग करके संभव नहीं है।

### 2.3 पायथन स्ट्रिंग को पायथन शब्दकोश सूची के रूप में पढ़ें

```python
import json 

def read_string_to_list(input_string):
    """
    将输入的字符串转换为 Python 列表。

    参数:
    input_string: 输入的字符串，应为有效的 JSON 格式。

    返回:
    list 或 None: 如果输入字符串有效，则返回对应的 Python 列表，否则返回 None。
    """
    if input_string is None:
        return None

    try:
# JSON प्रारूप की आवश्यकताओं को पूरा करने के लिए इनपुट स्ट्रिंग में सिंगल कोट्स को डबल कोट्स से बदलें
        input_string = input_string.replace("'", "\"")  
        data = json.loads(input_string)
        return data
    except json.JSONDecodeError:
        print("Error: Invalid JSON string")
        return None   
```

#### 2.3.1 वापस बुलाए गए उत्पादों और श्रेणियों का विवरण

```python
def generate_output_string(data_list):
    """
    根据输入的数据列表生成包含产品或类别信息的字符串。

    参数:
    data_list: 包含字典的列表，每个字典都应包含 "products" 或 "category" 的键。

    返回:
    output_string: 包含产品或类别信息的字符串。
    """
    if data_list is None:
        return output_string

    for data in data_list:
        try:
            if "products" in data:
                products_list = data["products"]
                for product_name in products_list:
                    product = get_product_by_name(product_name)
                    if product:
                        output_string += json.dumps(product, indent=4) + "\n"
                    else:
                        print(f"Error: Product '{product_name}' not found")
            elif "category" in data:
                category_name = data["category"]
                category_products = get_products_by_category(category_name)
                for product in category_products:
                    output_string += json.dumps(product, indent=4) + "\n"
            else:
                print("Error: Invalid object format")
        except Exception as e:
            print(f"Error: {e}")

    return output_string 
```

इसके अनेक कारण हैं।

सबसे पहले, बहुत सारे उत्पाद विवरण शामिल करने से संदर्भ को संसाधित करते समय मॉडल भ्रमित हो सकता है, ठीक वैसे ही जैसे किसी इंसान के लिए एक ही बार में बड़ी मात्रा में जानकारी संसाधित करने की कोशिश करना। बेशक, GPT-4 जैसे अधिक उन्नत मॉडलों के लिए, यह कारण कम महत्वपूर्ण है। विशेष रूप से जब संदर्भ इस उदाहरण की तरह अच्छी तरह से संरचित है, तो मॉडल स्पष्ट रूप से अप्रासंगिक जानकारी को चतुराई से अनदेखा करने के लिए पर्याप्त स्मार्ट है।

अगला कारण और भी अधिक ठोस है.

सबसे पहले, सभी उत्पाद विवरणों को शामिल करने से मॉडल अधिक प्रासंगिक रूप से अव्यवस्थित हो सकता है, क्योंकि यह एक बार में बड़ी मात्रा में जानकारी संसाधित करने की कोशिश करने वाले किसी व्यक्ति के लिए होगा। बेशक, यह प्रश्न GPT-4 जैसे अधिक उन्नत मॉडल के लिए कम प्रासंगिक है, खासकर जब संदर्भ इस उदाहरण की तरह अच्छी तरह से संरचित है और मॉडल स्पष्ट रूप से अप्रासंगिक जानकारी को अनदेखा करने के लिए पर्याप्त स्मार्ट है। अगला कारण अधिक ठोस है.

दूसरा कारण यह है कि भाषा मॉडल में संदर्भ बाधाएं होती हैं, यानी इनपुट और आउटपुट के रूप में निश्चित संख्या में टोकन की अनुमति होती है। यदि आपके पास एक विशाल उत्पाद सूची है, तो आप सभी विवरणों को संदर्भ विंडो में भी फिट नहीं कर सकते हैं।

अंतिम कारण यह है कि सभी उत्पाद विवरण शामिल करने से मॉडल पर अधिक प्रभाव पड़ सकता हैक्योंकि यह सभी उत्पाद विवरणों को याद रखता है, न कि केवल क्वेरी से संबंधित जानकारी को। इससे नए प्रश्नों को संभालते समय मॉडल खराब प्रदर्शन कर सकता है।

भाषा मॉडल का उपयोग करते समय, प्रति टोकन भुगतान के कारण यह महंगा हो सकता है। इसलिए, चुनिंदा जानकारी लोड करके, प्रतिक्रिया उत्पन्न करने की लागत को कम किया जा सकता है। सामान्य तौर पर, यह निर्धारित करना कि किसी मॉडल के संदर्भ में जानकारी को गतिशील रूप से कब लोड करना है, और मॉडल को यह निर्णय लेने की अनुमति देना कि कब अधिक जानकारी की आवश्यकता है, इन मॉडलों की क्षमताओं को बढ़ाने के सर्वोत्तम तरीकों में से एक है।

और फिर, आपको भाषा मॉडल को तर्क एजेंटों के रूप में सोचना चाहिए जिन्हें उपयोगी निष्कर्ष निकालने और उपयोगी कार्य करने के लिए आवश्यक संदर्भ की आवश्यकता होती है। इसलिए, इस मामले में, हमें मॉडल को उत्पाद जानकारी प्रदान करनी होगी, और फिर यह उपयोगकर्ता के लिए उपयोगी उत्तर बनाने के लिए उस उत्पाद जानकारी के आधार पर तर्क कर सकता है।

इस उदाहरण में, हमने उत्पाद के नाम से उत्पाद विवरण या श्रेणी के नाम से श्रेणी उत्पाद प्राप्त करने के लिए किसी फ़ंक्शन में एक विशिष्ट फ़ंक्शन या कॉल जोड़ा है। हालाँकि, मॉडल वास्तव में यह तय करने में अच्छे हैं कि विभिन्न उपकरणों का उपयोग कब करना है और उनका सही तरीके से उपयोग करना है। चैटजीपीटी प्लगइन के पीछे यही विचार है। हम मॉडल को बताते हैं कि उसके पास किन उपकरणों तक पहुंच है और वे क्या करते हैं, और वह उनका उपयोग तब करना चुनता है जब उसे किसी विशिष्ट स्रोत से जानकारी प्राप्त करने की आवश्यकता होती है या अन्य उचित कार्रवाई करना चाहता है। इस उदाहरण में, हम केवल सटीक उत्पाद और श्रेणी नाम मिलान के माध्यम से जानकारी पा सकते हैं, लेकिन अधिक उन्नत सूचना पुनर्प्राप्ति तकनीकें हैं। जानकारी प्राप्त करने के सबसे प्रभावी तरीकों में से एक प्राकृतिक भाषा प्रसंस्करण तकनीकों का उपयोग करना हैनाममात्र इकाई मान्यता और संबंध निष्कर्षण।

एक अन्य विधि जानकारी प्राप्त करने के लिए टेक्स्ट एम्बेडिंग (एंबेडिंग) का उपयोग करना है। किसी दिए गए प्रश्न से संबंधित जानकारी खोजने के लिए बड़े कॉर्पोरा पर कुशल ज्ञान पुनर्प्राप्ति को सक्षम करने के लिए एंबेडिंग का उपयोग किया जा सकता है। टेक्स्ट एम्बेडिंग का उपयोग करने का एक मुख्य लाभ यह है कि वे अस्पष्ट या अर्थ संबंधी खोजों को सक्षम करते हैं, जो आपको सटीक कीवर्ड का उपयोग किए बिना प्रासंगिक जानकारी ढूंढने की अनुमति देते हैं। इसलिए इस उदाहरण में हमें उत्पाद के सटीक नाम की आवश्यकता नहीं है, लेकिन "मोबाइल फ़ोन" जैसी अधिक सामान्य क्वेरी का उपयोग करके खोज सकते हैं। हम जल्द ही विभिन्न अनुप्रयोगों में एम्बेडिंग का उपयोग करने पर एक व्यापक पाठ्यक्रम शुरू करने की योजना बना रहे हैं, इसलिए बने रहें।

अगले अध्याय में हम चर्चा करेंगे कि किसी भाषा मॉडल के आउटपुट का मूल्यांकन कैसे करें।
