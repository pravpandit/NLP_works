### 2. पिवोट_टेबल

`पिवोट` का उपयोग विशिष्टता की स्थिति पर निर्भर करता है। यदि विशिष्टता की स्थिति पूरी नहीं होती है, तो एक ही पंक्ति और स्तंभ संयोजन के अनुरूप कई मानों को एकत्रीकरण संचालन के माध्यम से एक मान में बदल दिया जाना चाहिए। उदाहरण के लिए, झांग सैन और ली सी दोनों ने चीनी भाषा की परीक्षा और गणित की परीक्षा दो बार दी। कॉलेज के नियमों के अनुसार, अंतिम स्कोर दो टेस्ट अंकों का औसत है, इसे 'पिवोट' के माध्यम से पूरा नहीं किया जा सकता है `फ़ंक्शन.

```python
df = pd.DataFrame({'Name':['San Zhang', 'San Zhang', 
                              'San Zhang', 'San Zhang',
                              'Si Li', 'Si Li', 'Si Li', 'Si Li'],
                   'Subject':['Chinese', 'Chinese', 'Math', 'Math',
                                 'Chinese', 'Chinese', 'Math', 'Math'],
                   'Grade':[80, 90, 100, 90, 70, 80, 85, 95]})
df
```

`pandas` कार्यान्वयन के लिए `pivot_table` प्रदान करता है, और `aggfunc` पैरामीटर उपयोग किया गया समग्र फ़ंक्शन है। उपरोक्त परिदृश्य को इस प्रकार लिखा जा सकता है:

```python
df.pivot_table(index = 'Name',
               columns = 'Subject',
               values = 'Grade',
               aggfunc = 'mean')
```

यहां पारित `एगफंक` में पिछले अध्याय में प्रस्तुत सभी कानूनी एकत्रीकरण स्ट्रिंग्स शामिल हैं, इसके अलावा, आप एकत्रीकरण फ़ंक्शन को भी पास कर सकते हैं जो कस्टम संचालन को लागू करने के लिए इनपुट स्केलर के रूप में अनुक्रम लेता है समान रूप से लिखा गया:

```python
df.pivot_table(index = 'Name',
               columns = 'Subject',
               values = 'Grade',
               aggfunc = lambda x:x.mean())
```

इसके अलावा, `pivot_table` में सीमांत एकत्रीकरण का कार्य है, जिसे `margins=True` सेट करके प्राप्त किया जा सकता है, जहां सीमांत एकत्रीकरण विधि `aggfunc` में दी गई एकत्रीकरण विधि के अनुरूप है। निम्नलिखित आँकड़ों की गणना क्रमशः चीनी भाषा और गणित के औसत स्कोर, झांग सैन और ली सी के औसत स्कोर, साथ ही कुल मिलाकर सभी अंकों के औसत स्कोर के लिए की जाती है:

```python
df.pivot_table(index = 'Name',
               columns = 'Subject',
               values = 'Grade',
               aggfunc='mean',
               margins=True)
```

### 3.पिघलना

लंबी और चौड़ी तालिकाएँ केवल डेटा प्रस्तुत करने के तरीके में भिन्न होती हैं, लेकिन उनमें मौजूद जानकारी की मात्रा बराबर होती है, जैसा कि पहले उल्लेख किया गया है, लंबी तालिका को विस्तृत तालिका में बदलने के लिए `पिवोट` का उपयोग करें, फिर आप संबंधित रिवर्स का उपयोग कर सकते हैं चौड़ी टेबल को लंबी टेबल में बदलने का ऑपरेशन, `पिघल` फ़ंक्शन ऐसी भूमिका निभाता है। नीचे दिए गए उदाहरण में, `विषय` को एक कॉलम इंडेक्स के रूप में संग्रहीत किया गया है और अब आप इसे एक कॉलम में संपीड़ित करना चाहते हैं।

```python
df = pd.DataFrame({'Class':[1,2],
                   'Name':['San Zhang', 'Si Li'],
                   'Chinese':[80, 90],
                   'Math':[80, 75]})
df
```

```python
df_melted = df.melt(id_vars = ['Class', 'Name'],
                    value_vars = ['Chinese', 'Math'],
                    var_name = 'Subject',
                    value_name = 'Grade')
df_melted
```

`पिघल` और संपीड़न प्रक्रिया के मुख्य पैरामीटर नीचे दिए गए चित्र में दिखाए गए हैं:

<img src='../source/_static/ch5_melt.png' चौड़ाई='35%'>

जैसा कि पहले उल्लेख किया गया है, `पिघल` और `पिवट` पारस्परिक प्रक्रियाओं का एक सेट है, इसलिए आप निश्चित रूप से `df_melted` को `पिवट` ऑपरेशन के माध्यम से `df` के रूप में वापस परिवर्तित कर सकते हैं:

```python
df_unmelted = df_melted.pivot(index = ['Class', 'Name'],
                              columns='Subject',
                              values='Grade')
df_unmelted # 下面需要恢复索引，并且重命名列索引名称
```

```python
df_unmelted = df_unmelted.reset_index().rename_axis(columns={'Subject':''})
df_unmelted.equals(df)
```

### 4. चौड़ा_से_लंबा

`पिघल` विधि में, कॉलम इंडेक्स में संपीड़ित मानों के एक सेट के अनुरूप कॉलम तत्व केवल अर्थ के समान स्तर का प्रतिनिधित्व कर सकते हैं, अर्थात, `वैल्यू_नाम`। अब यदि कॉलम में क्रॉस-श्रेणियां हैं, जैसे कि मध्यावधि और अंतिम श्रेणियां और चीनी और गणित श्रेणियां, तो आप `values_name` के अनुरूप `ग्रेड` को चीनी अंकों और गणित के अंकों के अनुरूप दो कॉलम में विस्तारित करना चाहते हैं, और केवल मध्यावधि और अंतिम संपीड़न शामिल करें, इस मामले में, इसे पूरा करने के लिए `वाइड_टू_लॉन्ग` फ़ंक्शन का उपयोग किया जाना चाहिए।

```python
df = pd.DataFrame({'Class':[1,2],'Name':['San Zhang', 'Si Li'],
                   'Chinese_Mid':[80, 75], 'Math_Mid':[90, 85],
                   'Chinese_Final':[80, 75], 'Math_Final':[90, 85]})
df
```

```python
pd.wide_to_long(df,
                stubnames=['Chinese', 'Math'],
                i = ['Class', 'Name'],
                j='Examination',
                sep='_',
                suffix='.+')
```

विशिष्ट परिवर्तन प्रक्रिया को नीचे दिए गए चित्र में दिखाया गया है। समान अवधारणा से संबंधित तत्वों को सुसंगत रंगों से चिह्नित किया गया है:

<img src='../source/_static/ch5_wtl.png' चौड़ाई='35%'>

एक अधिक जटिल मामला नीचे दिया गया है। `पिवोट` अनुभाग (जिसने एक बहु-स्तरीय सूचकांक तैयार किया) में मल्टी-कॉलम ऑपरेशन का परिणाम `वाइड_टू_लॉन्ग` फ़ंक्शन का उपयोग करके अपने मूल रूप में परिवर्तित किया गया है। उनमें से, अध्याय 8 में `str.split` फ़ंक्शन का उपयोग किया जाता है, अभी के लिए, इसे केवल एक निश्चित विभाजक के अनुसार अनुक्रम को विभाजित करने के रूप में समझा जाना चाहिए।

```python
res = pivot_multi.copy()
res.columns = res.columns.map(lambda x:'_'.join(x))
res = res.reset_index()
res = pd.wide_to_long(res, stubnames=['Grade', 'rank'],
                           i = ['Class', 'Name'],
                           j = 'Subject_Examination',
                           sep = '_',
                           suffix = '.+')
res
```

```python
res = res.reset_index()
res[['Subject', 'Examination']] = res['Subject_Examination'].str.split('_', expand=True)
res = res[['Class', 'Name', 'Examination', 'Subject', 'Grade', 'rank']].sort_values('Subject')
res = res.reset_index(drop=True)
res
```

## 2. सूचकांक की विकृति

### 1. स्टैक और अनस्टैक

अध्याय 2 में, हमने सूचकांक के भीतर परत विनिमय के लिए `swaplevel` या `reorder_levels` के उपयोग का उल्लेख किया है। पंक्ति और स्तंभ अनुक्रमणिका के बीच $\color{लाल}$ के आदान-प्रदान पर नीचे चर्चा की जाएगी इसलिए डेटाफ़्रेम के आयामों में परिवर्तन विरूपण ऑपरेशन से संबंधित हैं। पहले खंड में उल्लिखित चार विरूपण कार्यों के बीच अंतर यह है कि वे सभी $\color{red}{element}$ और $\color{red}{column Index}$ रूपांतरण के बीच एक निश्चित कॉलम या कई कॉलम से संबंधित हैं, बल्कि सूचकांकों के बीच रूपांतरण की तुलना में।

`अनस्टैक` फ़ंक्शन का कार्य पंक्ति इंडेक्स को कॉलम इंडेक्स में परिवर्तित करना है, जैसे कि निम्नलिखित सरल उदाहरण:

```python
df = pd.DataFrame(np.ones((4,2)),
                  index = pd.Index([('A', 'cat', 'big'),
                                    ('A', 'dog', 'small'),
                                    ('B', 'cat', 'big'),
                                    ('B', 'dog', 'small')]),
                  columns=['col_1', 'col_2'])
df
```

```python
df.unstack()
```

`अनस्टैक` का मुख्य पैरामीटर चलती परत संख्या है। डिफ़ॉल्ट रूप से, सबसे भीतरी परत को परिवर्तित किया जाता है और कॉलम इंडेक्स की सबसे निचली परत में ले जाया जाता है। यह एक ही समय में कई परतों को परिवर्तित करने का भी समर्थन करता है।

```python
df.unstack(2)
```

```python
df.unstack([0,2])
```

`पिवोट` में विशिष्टता की आवश्यकता के समान, `अनस्टैक` में यह सुनिश्चित करना आवश्यक है कि $\color{red}{पंक्ति सूचकांक परत कॉलम इंडेक्स में परिवर्तित हो जाए}$ और $\color{red}{पंक्ति सूचकांक परत बरकरार रहे संयोजन }$ द्वारा गठित अद्वितीय है। उदाहरण के लिए, पहले दो कॉलम इंडेक्स को उसी में बदलने से विशिष्टता नष्ट हो जाती है, और एक त्रुटि रिपोर्ट की जाएगी:

```python
my_index = df.index.to_list()
my_index[1] = my_index[0]
df.index = pd.Index(my_index)
df
```

```python
try:
    df.unstack()
except Exception as e:
    Err_Msg = e
Err_Msg
```

`अनस्टैक` के विपरीत, `स्टैक` का कार्य कॉलम इंडेक्स को पंक्ति इंडेक्स में स्टैक करना है, और इसका उपयोग बिल्कुल समान है।

```python
df = pd.DataFrame(np.ones((4,2)),
                  index = pd.Index([('A', 'cat', 'big'),
                                    ('A', 'dog', 'small'),
                                    ('B', 'cat', 'big'),
                                    ('B', 'dog', 'small')]),
                  columns=['index_1', 'index_2']).T
df
```

```python
df.stack()
```

### 2. एकत्रीकरण और विरूपण के बीच संबंध

ऊपर पेश किए गए सभी कार्यों में, एकत्रीकरण प्रभाव वाले `pivot_table` को छोड़कर, सभी फ़ंक्शन परिवर्तन से पहले और बाद में `मानों` की संख्या को नहीं बदलेंगे, लेकिन इन मूल्यों की प्रस्तुति का रूप बदल जाएगा। पिछले अध्याय में चर्चा किया गया समूह एकत्रीकरण ऑपरेशन भी एक विशेष परिवर्तन ऑपरेशन होना चाहिए क्योंकि यह एक नई पंक्ति और स्तंभ सूचकांक उत्पन्न करता है, क्योंकि मूल एकाधिक मान एकत्रीकरण के बाद एक मान में बदल जाते हैं, `के व्यक्तिगत मान values` संख्या बदल गई है, जो समूहीकृत एकत्रीकरण और परिवर्तन फ़ंक्शन के बीच सबसे बड़ा अंतर भी है।

## 3. अन्य विरूपण कार्य

### 1. क्रॉसटैब

`क्रॉसस्टैब` एक अजीब फ़ंक्शन है, क्योंकि यह `पिवोट_टेबल` के सभी कार्यों को प्राप्त कर सकता है। डिफ़ॉल्ट स्थिति में, `क्रॉसस्टैब` तत्व संयोजनों की घटना की आवृत्ति की गणना कर सकता है, यानी, `गिनती` ऑपरेशन। उदाहरण के लिए, `learn_pandas` डेटा सेट में स्कूलों और विभाग के स्थानांतरण के अनुरूप आवृत्तियों की गणना करें:

```python
df = pd.read_csv('../data/learn_pandas.csv')
pd.crosstab(index = df.School, columns = df.Transfer)
```

यह `क्रॉसस्टैब` के निम्नलिखित लेखन के बराबर है, जहां `एगफंक` एकत्रीकरण पैरामीटर है:

```python
pd.crosstab(index = df.School, columns = df.Transfer, values = [0]*df.shape[0], aggfunc = 'count')
```

इसी तरह, आप समतुल्य संचालन करने के लिए `pivot_table` का उपयोग कर सकते हैं क्योंकि यहां आँकड़े संयोजनों की आवृत्तियाँ हैं, `मान` पैरामीटर अंतिम परिणाम को प्रभावित नहीं करेगा, चाहे कोई भी कॉलम पारित किया गया हो:

```python
df.pivot_table(index = 'School',
               columns = 'Transfer',
               values = 'Name',
               aggfunc = 'count')
```

जैसा कि ऊपर से देखा जा सकता है, दोनों कार्यों के बीच अंतर यह है कि `क्रॉसस्टैब` की संबंधित स्थिति को एक विशिष्ट अनुक्रम में पारित किया जाता है, जबकि `पिवोट_टेबल` को बुलाए गए तालिका के अनुरूप नाम में पारित किया जाता है अनुक्रम पारित हो गया है, एक त्रुटि की सूचना दी जाएगी।

डिफ़ॉल्ट स्थिति में `गिनती` आँकड़ों के अलावा, सभी एकत्रित स्ट्रिंग और कस्टम फ़ंक्शंस जो स्केलर लौटाते हैं, उपलब्ध हैं, जैसे संबंधित संयोजन की औसत ऊंचाई की गणना करना:

```python
pd.crosstab(index = df.School, columns = df.Transfer, values = df.Height, aggfunc = 'mean')
```
