[ए] स्ट्रिंग इंडेक्स के साथ `श्रृंखला`

यदि आप किसी एकल सूचकांक के संगत तत्व को निकालना चाहते हैं, तो आप `[आइटम]` का उपयोग कर सकते हैं। यदि `श्रृंखला` में केवल एक ही संगत मान है, तो स्केलर मान लौटाया जाता है। यदि कई संगत मान हैं, तो एक `श्रृंखला ` लौटाया गया है:

```python
s = pd.Series([1, 2, 3, 4, 5, 6], index=['a', 'b', 'a', 'a', 'a', 'c'])
s['a']
```

```python
s['b']
```

यदि आप एकाधिक सूचकांकों के संबंधित तत्वों को पुनः प्राप्त करना चाहते हैं, तो आप `[वस्तुओं की सूची]` का उपयोग कर सकते हैं:

```python
s[['c', 'b']]
```

यदि आप दो इंडेक्स के बीच के तत्वों को निकालना चाहते हैं, और ये दोनों इंडेक्स पूरे इंडेक्स में अद्वितीय हैं, तो आप एक स्लाइस का उपयोग कर सकते हैं, और ध्यान दें कि यहां स्लाइस में दो एंडपॉइंट होंगे:

```python
s['c': 'b': -2]
```

यदि आगे और पीछे के समापन बिंदुओं के मान दोहराए जाते हैं, तो स्लाइस का उपयोग करने से पहले उन्हें क्रमबद्ध करने की आवश्यकता होती है:

```python
try:
    s['a': 'b']
except Exception as e:
    Err_Msg = e
Err_Msg
```

```python
s.sort_index()['a': 'b']
```

[बी] पूर्णांक सूचकांक के साथ `श्रृंखला`

डेटा रीडिंग फ़ंक्शन का उपयोग करते समय, यदि संबंधित कॉलम को इंडेक्स के रूप में निर्दिष्ट नहीं किया जाता है, तो 0 से शुरू होने वाला एक पूर्णांक इंडेक्स डिफ़ॉल्ट इंडेक्स के रूप में उत्पन्न होगा, लंबाई आवश्यकताओं को पूरा करने वाले पूर्णांकों का कोई भी सेट इंडेक्स के रूप में उपयोग किया जा सकता है .

स्ट्रिंग्स की तरह, यदि `[int]` या `[int_list]` का उपयोग किया जाता है, तो संबंधित सूचकांक **तत्व** का मान पुनर्प्राप्त किया जा सकता है:

```python
s = pd.Series(['a', 'b', 'c', 'd', 'e', 'f'], index=[1, 3, 1, 2, 5, 4])
s[1]
```

```python
s[[2,3]]
```

यदि आप पूर्णांक स्लाइसिंग का उपयोग करते हैं, तो संबंधित सूचकांक **स्थिति** का मान निकाल लिया जाएगा। ध्यान दें कि यहां पूर्णांक स्लाइसिंग में सही समापन बिंदु शामिल नहीं है, ठीक `पायथन` में स्लाइसिंग की तरह:

```python
s[1:-1:2]
```

### 3.लोक इंडेक्सर

हमने पहले 'डेटाफ़्रेम' के कॉलम चुनने के बारे में बात की थी, और अब हम टेबल के लिए इसकी पंक्ति चयन पर चर्चा करेंगेतों, दो इंडेक्सर हैं, एक **तत्व**-आधारित `loc` इंडेक्सर है, और दूसरा **स्थिति**-आधारित `iloc` इंडेक्सर है।

`loc` इंडेक्सर का सामान्य रूप `loc[*, *]` है, जहां पहला `*` पंक्तियों के चयन का प्रतिनिधित्व करता है, और दूसरा `*` कॉलम के चयन का प्रतिनिधित्व करता है यदि दूसरी स्थिति छोड़ दी जाती है `loc[*]` के रूप में लिखा गया है, यह `*` पंक्तियों के चयन को संदर्भित करता है, उनमें से `*` स्थिति में पांच प्रकार की कानूनी वस्तुएं हैं, अर्थात्: एकल तत्व, तत्व लिसटी, एलिमेंट स्लाइस, बूलियन सूची और फ़ंक्शन, जिसे नीचे बारी-बारी से समझाया जाएगा।

संबंधित ऑपरेशन को प्रदर्शित करने के लिए, हम पहले `नाम` कॉलम को एक इंडेक्स के रूप में सेट करने के लिए `set_index` विधि का उपयोग करते हैं। इस फ़ंक्शन के अन्य उपयोग मल्टी-लेवल इंडेक्सिंग पर अध्याय में पेश किए जाएंगे।

```python
df_demo = df.set_index('Name')
df_demo.head()
```

[डी] `*` एक बूलियन सूची है

वास्तविक डेटा प्रोसेसिंग में, शर्तों के आधार पर पंक्तियों को फ़िल्टर करना बहुत आम है, यहां `loc` को पास की गई बूलियन सूची की लंबाई `DataFrame` के समान है, और उन स्थानों के अनुरूप पंक्तियों की लंबाई `True` होगी। चयनित किया जाएगा, और `गलत` हटा दिया जाएगा।

उदाहरण के लिए, 70 किलोग्राम से अधिक वजन वाले छात्रों का चयन करें:

```python
df_demo.loc[df_demo.Weight>70].head()
```

ऊपर उल्लिखित पारित तत्वों की सूची को 'isin' विधि द्वारा लौटाई गई बूलियन सूची का उपयोग करके समकक्ष रूप से भी लिखा जा सकता है, उदाहरण के लिए, नए और वरिष्ठ नागरिकों की सभी जानकारी का चयन करने के लिए:

```python
df_demo.loc[df_demo.Grade.isin(['Freshman', 'Senior'])].head()
```

यौगिक स्थितियों के लिए, आप इसे प्राप्त करने के लिए | (या), और (और), और ~ (नकारात्मक) के संयोजन का उपयोग कर सकते हैं, उदाहरण के लिए, फुडन विश्वविद्यालय में वरिष्ठ छात्रों का चयन करें जिनका वजन 70 किलोग्राम से अधिक है, या पुरुष छात्र हैं पेकिंग विश्वविद्यालय जिनका वजन 80 किलोग्राम से अधिक है लेकिन वे वरिष्ठ नहीं हैं:

```python
condition_1_1 = df_demo.School == 'Fudan University'
condition_1_2 = df_demo.Grade == 'Senior'
condition_1_3 = df_demo.Weight > 70
condition_1 = condition_1_1 & condition_1_2 & condition_1_3
condition_2_1 = df_demo.School == 'Peking University'
condition_2_2 = df_demo.Grade == 'Senior'
condition_2_3 = df_demo.Weight > 80
condition_2 = condition_2_1 & (~condition_2_2) & condition_2_3
df_demo.loc[condition_1 | condition_2]
```

【e】`*` एक फ़ंक्शन है

यहां फ़ंक्शन को ऊपर उल्लिखित चार कानूनी रूपों में से एक को वापस करना होगा, और फ़ंक्शन का इनपुट मान `DataFr' हैame` स्वयं। यह मानते हुए कि यह अभी भी ऊपर उल्लिखित यौगिक स्थिति स्क्रीनिंग का उदाहरण है, तर्क को एक फ़ंक्शन में लिखा जा सकता है और फिर वापस किया जा सकता है। यह ध्यान दिया जाना चाहिए कि फ़ंक्शन का औपचारिक पैरामीटर `x` अनिवार्य रूप से `df_demo` है।

```python
def condition(x):
    condition_1_1 = x.School == 'Fudan University'
    condition_1_2 = x.Grade == 'Senior'
    condition_1_3 = x.Weight > 70
    condition_1 = condition_1_1 & condition_1_2 & condition_1_3
    condition_2_1 = x.School == 'Peking University'
    condition_2_2 = x.Grade == 'Senior'
    condition_2_3 = x.Weight > 80
    condition_2 = condition_2_1 & (~condition_2_2) & condition_2_3
    result = condition_1 | condition_2
    return result
df_demo.loc[condition]
```

इसके अलावा, लैम्ब्डा एक्सप्रेशन भी समर्थित हैं, और उनका रिटर्न मान भी पहले बताए गए चार रूपों में से एक होना चाहिए:

```python
df_demo.loc[lambda x:'Quan Zhao', lambda x:'Gender']
```

चूंकि फ़ंक्शन `स्टार्ट: एंड: स्टेप` जैसे स्लाइस प्रारूप को वापस नहीं कर सकता है, इसलिए इसे वापस करते समय स्लाइस को `स्लाइस` ऑब्जेक्ट के साथ लपेटा जाना चाहिए:

```python
df_demo.loc[lambda x: slice('Gaojuan You', 'Gaoqiang Qian')]
```

अंत में, यह बताया जाना चाहिए कि लोक इंडेक्स का उपयोग श्रृंखला के लिए भी किया जा सकता है। अपनाया गया सिद्धांत बिल्कुल वैसा ही है जैसा कि डेटाफ़्रेम में पंक्ति फ़िल्टरिंग के लिए उपयोग किया जाने वाला लोक [*] है, इसलिए मैं यहां विवरण में नहीं जाऊंगा।

#### 【चेतावनी】जंजीरयुक्त असाइनमेंट का उपयोग न करें

किसी तालिका या अनुक्रम में मान निर्दिष्ट करते समय, आपको इंडेक्सर्स की एक परत का उपयोग करने के बाद सीधे असाइनमेंट ऑपरेशन करना चाहिए, ऐसा इसलिए है क्योंकि एकाधिक इंडेक्स के बाद, असाइनमेंट अस्थायी रूप से लौटाई गई 'कॉपी' कॉपी और तत्व को सौंपा जाता है वास्तव में संशोधित नहीं किया गया है, इस प्रकार 'SettingWithCopyWarning' चेतावनी की रिपोर्ट की गई है, उदाहरण के लिए:

```python
df_chain = pd.DataFrame([[0,0],[1,0],[-1,0]], columns=list('AB'))
df_chain
import warnings
with warnings.catch_warnings():
    warnings.filterwarnings('error')
    try:
        df_chain[df_chain.A!=0].B = 1 # 使用方括号列索引后，再使用点的列索引
    except Warning as w:
        Warning_Msg = w
print(Warning_Msg)
df_chain
```

```python
df_chain.loc[df_chain.A!=0,'B'] = 1
df_chain
```

### 6. यादृच्छिक नमूनाकरण

यदि आप 'डेटाफ़्रेम' की प्रत्येक पंक्ति को एक नमूना मानते हैं, या प्रत्येक कॉलम को एक सुविधा के रूप में मानते हैं, और संपूर्ण 'डेटाफ़्रेम' को एक जनसंख्या के रूप में मानते हैं, तो आप कभी-कभी बड़े पैमाने पर नमूने या सुविधाओं का नमूना लेने के लिए 'नमूना' फ़ंक्शन का उपयोग कर सकते हैं डेटा सेट, आप डेटा के अनुमानित वितरण को समझने के लिए सांख्यिकीय विशेषताओं की गणना करना चाहते हैं, लेकिन इसमें बहुत समय लगता है, क्योंकि कई सांख्यिकीय विशेषताएं सांख्यिकीय विशेषताओं के निष्पक्ष अनुमान हैंसमान संभावना और बिना किसी प्रतिस्थापन के सरल यादृच्छिक नमूने की स्थिति के तहत जनसंख्या का, जैसे नमूना माध्य और जनसंख्या माध्य, आप अनुमानित अनुमान लगाने के लिए पहले पूरी तालिका से एक हिस्सा निकाल सकते हैं।

`नमूना` फ़ंक्शन में मुख्य पैरामीटर `एन, अक्ष, फ़्रेक, प्रतिस्थापित, वजन` हैं। पहले तीन नमूनों की संख्या, नमूने की दिशा (पंक्तियों के लिए 0 और स्तंभों के लिए 1), और नमूना अनुपात को संदर्भित करते हैं। (0.3 का मतलब है कि 30% नमूने लिए गए हैंआबादी)।

`रिप्लेस` और `वेट` क्रमशः प्रत्येक नमूने को प्रतिस्थापित करने और नमूना लेने की सापेक्ष संभावना को संदर्भित करते हैं। जब `रिप्लेस = ट्रू` होता है, तो इसका मतलब प्रतिस्थापन के साथ नमूनाकरण होता है, उदाहरण के लिए, निम्नलिखित निर्मित `df_sample` के सापेक्ष आकार का उपयोग करता है प्रतिस्थापन के साथ नमूनाकरण करने की नमूना संभावना के रूप में `मूल्य`, और नमूनों की संख्या 3 है।

```python
df_sample = pd.DataFrame({'id': list('abcde'), 'value': [1, 2, 3, 4, 90]})
df_sample
```

```python
df_sample.sample(3, replace = True, weights = df_sample.value)
```

### 3. इंडेक्सस्लाइस ऑब्जेक्ट

ऊपर प्रस्तुत विधि केवल टुपल को संपूर्ण रूप से स्लाइस कर सकती है, लेकिन प्रत्येक परत को नहीं, भले ही इंडेक्स दोहराया न जाए। इसे बूलियन सूचियों के साथ स्लाइस को मिलाने की भी अनुमति नहीं है। `इंडेक्सस्लाइस` ऑब्जेक्ट की शुरूआत इस समस्या को हल कर सकती है `स्लाइस` ऑब्जेक्ट के दो रूप हैं, पहला `loc[idx[*,*]]` प्रकार है, और दूसरा `loc[idx[*,*],idx[*,*]]` है। प्रकार, जिसे नीचे प्रस्तुत किया जाएगा प्रदर्शन की सुविधा के लिए, निम्नलिखित एक **दा का निर्माण करता हैtaFrame` गैर-डुप्लिकेट अनुक्रमणिका के साथ:

```python
np.random.seed(0)
L1,L2 = ['A','B','C'],['a','b','c']
mul_index1 = pd.MultiIndex.from_product([L1,L2],names=('Upper', 'Lower'))
L3,L4 = ['D','E','F'],['d','e','f']
mul_index2 = pd.MultiIndex.from_product([L3,L4],names=('Big', 'Small'))
df_ex = pd.DataFrame(np.random.randint(-9,10,(9,9)), index=mul_index1, columns=mul_index2)
df_ex
```

साइल ऑब्जेक्ट का उपयोग करने के लिए, आपको पहले इसे परिभाषित करना होगा:

```python
idx = pd.IndexSlice
```

[ए] `loc[idx[*,*]]` प्रकार

इस मामले में, स्लाइसिंग की कई परतें अलग-अलग नहीं की जा सकतीं। पहला `*` पंक्तियों के चयन को इंगित करता है, और दूसरा `*` स्तंभों के चयन को इंगित करता है, जो सरल `loc` के समान है:

```python
df_ex.loc[idx['C':, ('D', 'f'):]]
```

इसके अतिरिक्त, बूलियन अनुक्रमों का अनुक्रमण समर्थित है:

```python
df_ex.loc[idx[:'A', lambda x:x.sum()>0]] # 列和大于0
```

【b】`loc[idx[*,*],idx[*,*]]` प्रकार

इस मामले में, स्लाइसिंग परतों में की जा सकती है, जिसमें पहला `idx` पंक्ति सूचकांक को संदर्भित करता है और दूसरा स्तंभ सूचकांक को संदर्भित करता है।

```python
df_ex.loc[idx[:'A', 'b':], idx['E':, 'e':]]
```

लेकिन कृपया ध्यान दें कि फ़ंक्शंस का उपयोग इस समय समर्थित नहीं है:

```python
try:
    df_ex.loc[idx[:'A', lambda x: 'b'], idx['E':, 'e':]]
except Exception as e:
    Err_Msg = e
Err_Msg
```

पारित पैरामीटर एक फ़ंक्शन भी हो सकता है, जिसका इनपुट मान सूचकांक तत्व है:

#### 【अभ्यास】

```python
new_values = iter(list('abcdefgh'))
df_ex.rename(index=lambda x:next(new_values), level=2)
```

यदि आप किसी तत्व को किसी निश्चित स्थिति में संशोधित करना चाहते हैं, तो एकल-स्तरीय सूचकांक में ऐसा करना आसान है, यानी पहले सूचकांक की `मान` विशेषता को हटा दें, फिर प्राप्त सूची को संशोधित करें, और अंत में `` को पुन: असाइन करें इंडेक्स` ऑब्जेक्ट। हालाँकि, यदि यह एक बहु-स्तरीय इंडेक्स है, तो यह थोड़ा परेशानी भरा होगा। एक समाधान यह है कि इंडेक्स के एक निश्चित स्तर को अस्थायी रूप से तालिका के एक तत्व में परिवर्तित किया जाए, फिर इसे संशोधित किया जाए और अंत में इसे रीसेट किया जाए अनुक्रमणिका। निम्नलिखित अनुभाग इन परिचालनों का परिचय देगा।एक अन्य फ़ंक्शन जिसे पेश करने की आवश्यकता है वह है `मैप`, जो `इंडेक्स` पर परिभाषित एक विधि है, यह पिछली `रीनेम` विधि में परत के कार्यात्मक उपयोग के समान है, सिवाय इसके कि यह स्केलर मान में पास नहीं होता है। परत का, लेकिन सीधे सूचकांक के टुपल में गुजरता है, जो उपयोगकर्ताओं को क्रॉस-स्तरीय संशोधन करने के लिए ट्रैवर्सल प्रदान करता है। उदाहरण के लिए, उपरोक्त स्ट्रिंग रूपांतरण ऑपरेशन को समकक्ष रूप से लिखा जा सकता है:

```python
df_temp = df_ex.copy()
new_idx = df_temp.index.map(lambda x: (x[0], x[1], str.upper(x[2])))
df_temp.index = new_idx
df_temp.head()
```

`मैप` का एक अन्य उपयोग बहु-स्तरीय अनुक्रमणिका को संपीड़ित करना है, जो अध्याय 4 और अध्याय 5 में कुछ कार्यों में उपयोगी है:

```python
df_temp = df_ex.copy()
new_idx = df_temp.index.map(lambda x: (x[0]+'-'+x[1]+'-'+x[2]))
df_temp.index = new_idx
df_temp.head() # 单层索引
```

साथ ही इसे विपरीत दिशा में भी विस्तारित किया जा सकता है:

```python
new_idx = df_temp.index.map(lambda x:tuple(x.split('-')))
df_temp.index = new_idx
df_temp.head() # 三层索引
```

## 4. सूचकांक संचालन
### 1. संचालन नियम निर्धारित करें

शर्तों को पूरा करने वाली पंक्तियों को निकालने के लिए अक्सर सेट ऑपरेशंस का उपयोग करने की आवश्यकता होती है। उदाहरण के लिए, दो तालिकाएँ 'ए' और 'बी' हैं, और उनके सूचकांक कर्मचारी संख्याएँ हैं इस समय, दो तालिकाओं के अनुक्रमणिका को 'सूचकांक' पर संचालन के माध्यम से प्राप्त करना आसान है।

हालाँकि, उससे पहले, आइए चार सामान्य सेट ऑपरेशनों की समीक्षा करें:

$$\rm S_A.intersection(S_B) = \rm S_A \cap S_B \Leftrightarrow \rm \{x|x\in S_A\, और\, x\in S_B\}$$
$$\rm S_A.union(S_B) = \rm S_A \cup S_B \Leftrightarrow \rm \{x|x\in S_A\, या\, x\in S_B\}$$
$$\rm S_A.difference(S_B) = \rm S_A - S_B \Leftrightarrow \rm \{x|x\in S_A\, और\, x\notin S_B\}$$
$$\rm S_A.symmetric\_difference(S_B) = \rm S_A\triangle S_B\Leftrightarrow \rm \{x|x\in S_A\cup S_B - S_A\cap S_B\}$$

### 2. सामान्य सूचकांक संचालन

चूंकि संग्रह के तत्व अलग-अलग हैं, लेकिन सूचकांक में समान तत्व हो सकते हैं, ऑपरेशन करने से पहले डुप्लिकेट को हटाने के लिए `अद्वितीय` का उपयोग करें। निम्नलिखित प्रदर्शन के लिए दो सरलतम उदाहरण तालिकाएँ बनाता है:

```python
df_set_1 = pd.DataFrame([[0,1],[1,2],[3,4]], index = pd.Index(['a','b','a'],name='id1'))
df_set_2 = pd.DataFrame([[4,5],[2,6],[7,1]], index = pd.Index(['b','b','c'],name='id2'))
id1, id2 = df_set_1.index.unique(), df_set_2.index.unique()
id1.intersection(id2)
```

```python
id1.union(id2)
```

```python
id1.difference(id2)
```

```python
id1.symmetric_difference(id2)
```

यदि ऑपरेशन के लिए सेट किए जाने वाले दो तालिकाओं के कॉलम अनुक्रमित नहीं हैं, तो एक तरीका यह है कि पहले उन्हें इंडेक्स में परिवर्तित किया जाए, फिर ऑपरेशन के बाद उन्हें पुनर्स्थापित किया जाए। उदाहरण के लिए, `isin` फ़ंक्शन का उपयोग किया जाए पहली तालिका जहां सूचकांक रीसेट किया गया है, आईडी कॉलम के प्रतिच्छेदन वाली पंक्ति का चयन करें:

```python
df_set_in_col_1 = df_set_1.reset_index()
df_set_in_col_2 = df_set_2.reset_index()
df_set_in_col_1
```
