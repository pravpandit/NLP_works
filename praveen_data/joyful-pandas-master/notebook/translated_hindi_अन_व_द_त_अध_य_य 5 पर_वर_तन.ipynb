### 1.पिवोटी

`पिवोट` एक लंबी तालिका को एक विस्तृत तालिका में बदलने के लिए एक विशिष्ट कार्य है। सबसे पहले, आइए एक उदाहरण देखें: निम्नलिखित तालिका झांग सैन और ली सी के चीनी और गणित अंकों को संग्रहीत करती है। अब हम चीनी को प्रदर्शित करना चाहते हैं और गणित के अंक कॉलम के रूप में।

```python
df = pd.DataFrame({'Class':[1,1,2,2],
                   'Name':['San Zhang','San Zhang','Si Li','Si Li'],
                   'Subject':['Chinese','Math','Chinese','Math'],
                   'Grade':[80,75,90,85]})
df
```

बुनियादी लंबाई-से-चौड़ाई ऑपरेशन के लिए, तीन सबसे महत्वपूर्ण तत्व हैं रूपांतरित पंक्ति सूचकांक, स्तंभ को स्तंभ सूचकांक में परिवर्तित किया जाना है, और इन स्तंभ और पंक्ति सूचकांकों से संबंधित मान, जो इसके अनुरूप हैं `पिवट` विधि में `सूचकांक, कॉलम, मान` पैरामीटर। नई जेनरेट की गई तालिका का कॉलम सूचकांक `कॉलम` के संबंधित कॉलम का `अद्वितीय` मान है, और नई तालिका का पंक्ति सूचकांक `अद्वितीय` है `` के संबंधित कॉलम का मानdex`, और `मान` उस संख्यात्मक कॉलम से मेल खाते हैं जिसे आप प्रदर्शित करना चाहते हैं।

```python
df.pivot(index='Name', columns='Subject', values='Grade')
```

### 2. पिवोट_टेबल

`पिवोट` का उपयोग अद्वितीय पर निर्भर करता हैयदि विशिष्टता की स्थिति पूरी नहीं होती है, तो एक ही पंक्ति और स्तंभ संयोजन के अनुरूप कई मानों को एक मान बनने के लिए एकत्रित किया जाना चाहिए। उदाहरण के लिए, झांग सैन और ली सी दोनों ने दो चीनी और गणित परीक्षाएं दीं कॉलेज के नियमों के अनुसार, अंतिम स्कोर दो परीक्षा अंकों का औसत है, इस समय, इसे 'पिवोट' फ़ंक्शन के माध्यम से पूरा नहीं किया जा सकता है।

```python
df = pd.DataFrame({'Name':['San Zhang', 'San Zhang', 
                              'San Zhang', 'San Zhang',
                              'Si Li', 'Si Li', 'Si Li', 'Si Li'],
                   'Subject':['Chinese', 'Chinese', 'Math', 'Math',
                                 'Chinese', 'Chinese', 'Math', 'Math'],
                   'Grade':[80, 90, 100, 90, 70, 80, 85, 95]})
df
```

`पांडा` इसे प्राप्त करने के लिए `पिवोट_टेबल` प्रदान करता है, जहां `एगफंक` पैरामीटर उपयोग किया गया एकत्रीकरण फ़ंक्शन है। उपरोक्त परिदृश्य को निम्नानुसार लिखा जा सकता है:

```python
df.pivot_table(index = 'Name',
               columns = 'Subject',
               values = 'Grade',
               aggfunc = 'mean')
```

यहां पारित `एगफंक` में पिछले अध्याय में प्रस्तुत सभी कानूनी एकत्रीकरण स्ट्रिंग्स शामिल हैं, इसके अलावा, आप कस्टम संचालन को लागू करने के लिए इनपुट के रूप में एक अनुक्रम और आउटपुट के रूप में एक स्केलर के साथ एक एकत्रीकरण फ़ंक्शन भी पास कर सकते हैं समान रूप से इस प्रकार लिखा गया है:

```python
df.pivot_table(index = 'Name',
               columns = 'Subject',
               values = 'Grade',
               aggfunc = lambda x:x.mean())
```

इसके अलावा, `pivot_table` में सीमांत एकत्रीकरण का कार्य है, जिसे `margins=True` सेट करके प्राप्त किया जा सकता है, जहां सीमांत एकत्रीकरण विधि `aggfunc` में दी गई एकत्रीकरण विधि के अनुरूप है। निम्नलिखित आँकड़ों की क्रमशः गणना की जाती है चीनी और गणित के औसत अंक, झांग सैन और ली सी के औसत अंक, और कुल मिलाकर सभी अंकों का औसत अंक:

```python
df.pivot_table(index = 'Name',
               columns = 'Subject',
               values = 'Grade',
               aggfunc='mean',
               margins=True)
```

### 3.पिघलना

लंबी और चौड़ी तालिकाएँ डेटा प्रस्तुत करने के तरीके में भिन्न होती हैं, लेकिन उनमें मौजूद जानकारी की मात्रा `पिवोट` का उपयोग करने के बराबर होती हैएक लंबी तालिका को एक विस्तृत तालिका में परिवर्तित करने के लिए, संबंधित व्युत्क्रम ऑपरेशन का उपयोग एक विस्तृत तालिका को लंबी तालिका में परिवर्तित करने के लिए किया जा सकता है। निम्नलिखित उदाहरण में, `पिघल` फ़ंक्शन ऐसी भूमिका निभाता है कॉलम इंडेक्स का रूप, और अब हम इसे एक कॉलम में संपीड़ित करना चाहते हैं।

```python
df = pd.DataFrame({'Class':[1,2],
                   'Name':['San Zhang', 'Si Li'],
                   'Chinese':[80, 90],
                   'Math':[80, 75]})
df
```

```python
df_melted = df.melt(id_vars = ['Class', 'Name'],
                    value_vars = ['Chinese', 'Math'],
                    var_name = 'Subject',
                    value_name = 'Grade')
df_melted
```

`पिघल` और संपीड़न प्रक्रिया के मुख्य पैरामीटर नीचे दिए गए चित्र में दिखाए गए हैं:

<img src='../source/_static/ch5_melt.png' चौड़ाई='35%'>

जैसा कि पहले उल्लेख किया गया है, `पिघल` और `पिवट` व्युत्क्रम प्रक्रियाओं का एक सेट है, इसलिए यह निश्चित है कि `df_melted` को `पिवट` ऑपरेशन के माध्यम से वापस `df` में परिवर्तित किया जा सकता है:

```python
df_unmelted = df_melted.pivot(index = ['Class', 'Name'],
                              columns='Subject',
                              values='Grade')
df_unmelted # 下面需要恢复索引，并且重命名列索引名称
```

```python
df_unmelted = df_unmelted.reset_index().rename_axis(columns={'Subject':''})
df_unmelted.equals(df)
```

### 4. चौड़ा_से_लंबा

`पिघल` विधि में, कॉलम इंडेक्स में संपीड़ित मानों के एक सेट के अनुरूप कॉलम तत्व केवल अर्थ के समान स्तर का प्रतिनिधित्व कर सकते हैं, अर्थात, `वैल्यू_नाम`। अब यदि कॉलम में क्रॉस-श्रेणियां हैं, जैसे कि मध्यावधि और अंतिम परीक्षा की श्रेणियां और चीनी और गणित की श्रेणियां, तो यदि आप क्रमशः चीनी अंकों और गणित के अंकों के अनुरूप दो कॉलमों में `values_name` के अनुरूप `ग्रेड` का विस्तार करना चाहते हैं, और केवल संपीड़ित करते हैंयह मध्यावधि और अंतिम परीक्षाओं की जानकारी है, इस आवश्यकता को पूरा करने के लिए आपको `वाइड_टू_लॉन्ग` फ़ंक्शन का उपयोग करना होगा।

```python
df = pd.DataFrame({'Class':[1,2],'Name':['San Zhang', 'Si Li'],
                   'Chinese_Mid':[80, 75], 'Math_Mid':[90, 85],
                   'Chinese_Final':[80, 75], 'Math_Final':[90, 85]})
df
```

```python
pd.wide_to_long(df,
                stubnames=['Chinese', 'Math'],
                i = ['Class', 'Name'],
                j='Examination',
                sep='_',
                suffix='.+')
```

विशिष्ट परिवर्तन प्रक्रिया को नीचे दिए गए चित्र में दिखाया गया है। एक ही अवधारणा के तत्वों को एक ही रंग से चिह्नित किया गया है:

<img src='../source/_static/ch5_wtl.png' चौड़ाई='35%'>

निम्नलिखित एक अधिक जटिल उदाहरण है। पिछले अनुभाग `पिवोट` (बहु-स्तरीय सूचकांक उत्पन्न करना) में बहु-स्तंभ ऑपरेशन का परिणाम `वाइड_टू_लॉन्ग` फ़ंक्शन का उपयोग करके अपने मूल रूप में परिवर्तित हो जाता है अध्याय 8 में फ़ंक्शन का उपयोग किया गया है, अभी के लिए इसे अनुक्रम को विभाजित करने के रूप में समझा जा सकता हैई एक निश्चित सीमांकक के अनुसार.

```python
res = pivot_multi.copy()
res.columns = res.columns.map(lambda x:'_'.join(x))
res = res.reset_index()
res = pd.wide_to_long(res, stubnames=['Grade', 'rank'],
                           i = ['Class', 'Name'],
                           j = 'Subject_Examination',
                           sep = '_',
                           suffix = '.+')
res
```

```python
res = res.reset_index()
res[['Subject', 'Examination']] = res['Subject_Examination'].str.split('_', expand=True)
res = res[['Class', 'Name', 'Examination', 'Subject', 'Grade', 'rank']].sort_values('Subject')
res = res.reset_index(drop=True)
res
```

## 2. सूचकांक परिवर्तन

### 1. स्टैक और अनस्टैक

अध्याय 2 में, हमने एक सूचकांक के भीतर परतों को स्वैप करने के लिए `swaplevel` या `reorder_levels` का उपयोग करने का उल्लेख किया है, अब हम $\color{red}{row और column indices}$ के आदान-प्रदान पर चर्चा करेंगे 'डेटाफ़्रेम' के आयाम, यह एक परिवर्तन ऑपरेशन है। धारा 1 में उल्लिखित चार परिवर्तन कार्य इस मायने में भिन्न हैं कि वे सभी $\color{red}{Elements' के एक या अधिक स्तंभों के बीच रूपांतरण से संबंधित हैं।}$ और $\color{red}{column indices}$, बजाय सूचकांकों के बीच रूपांतरण के।

`अनस्टैक` फ़ंक्शन का कार्य पंक्ति सूचकांकों को स्तंभ सूचकांकों में परिवर्तित करना है, जैसे कि निम्नलिखित सरल उदाहरण:

```python
df = pd.DataFrame(np.ones((4,2)),
                  index = pd.Index([('A', 'cat', 'big'),
                                    ('A', 'dog', 'small'),
                                    ('B', 'cat', 'big'),
                                    ('B', 'dog', 'small')]),
                  columns=['col_1', 'col_2'])
df
```

```python
df.unstack()
```

`अनस्टैक` का मुख्य पैरामीटर स्थानांतरित की जाने वाली परत संख्या है। डिफ़ॉल्ट रूप से, सबसे भीतरी परत को परिवर्तित किया जाता है और कॉलम इंडेक्स की सबसे निचली परत में ले जाया जाता है। यह एक ही समय में कई परतों को परिवर्तित करने का भी समर्थन करता है।

```python
df.unstack(2)
```

```python
df.unstack([0,2])
```

`पिवोट` में विशिष्टता की आवश्यकता के समान, `अनस्टैक` में, यह सुनिश्चित करना आवश्यक है कि $\color{red}{row इंडेक्स लेयर का संयोजन कॉलम इंडेक्स}$ में परिवर्तित हो जाए और $\color{red}{पंक्ति बरकरार रहे इंडेक्स लेयर}$ अद्वितीय है। उदाहरण के लिए, पहले दो कॉलम इंडेक्स को उसी में बदलने से विशिष्टता नष्ट हो जाती है, और एक त्रुटि रिपोर्ट की जाएगी:

```python
my_index = df.index.to_list()
my_index[1] = my_index[0]
df.index = pd.Index(my_index)
df
```

```python
try:
    df.unstack()
except Exception as e:
    Err_Msg = e
Err_Msg
```

`अनस्टैक` के विपरीत, `स्टैक` का कार्य कॉलम इंडेक्स की परतों को पंक्ति इंडेक्स में स्टैक करना है, और इसका उपयोग बिल्कुल समान है।

```python
df = pd.DataFrame(np.ones((4,2)),
                  index = pd.Index([('A', 'cat', 'big'),
                                    ('A', 'dog', 'small'),
                                    ('B', 'cat', 'big'),
                                    ('B', 'dog', 'small')]),
                  columns=['index_1', 'index_2']).T
df
```

## 3. अन्य परिवर्तन कार्य

### 1. क्रॉसटैब

`क्रॉसस्टैब` एक अजीब फ़ंक्शन है क्योंकि यह जो भी कार्य प्राप्त कर सकता है वह `पिवोट_टेबल` द्वारा पूरा किया जा सकता है। डिफ़ॉल्ट स्थिति में, `क्रॉसस्टैब` तत्व संयोजनों की आवृत्ति की गणना कर सकता है, अर्थात, `काउंट` ऑपरेशन। सह`learn_pandas` डेटासेट में स्कूलों और स्थानांतरणों की आवृत्ति:

```python
df = pd.read_csv('../data/learn_pandas.csv')
pd.crosstab(index = df.School, columns = df.Transfer)
```

यह निम्नलिखित `क्रॉसस्टैब` के बराबर है, जहां `एगफंक` एकत्रीकरण पैरामीटर है:

```python
pd.crosstab(index = df.School, columns = df.Transfer, values = [0]*df.shape[0], aggfunc = 'count')
```

इसी तरह, आप समतुल्य संचालन करने के लिए `pivot_table` का उपयोग कर सकते हैं क्योंकि यहां संयोजनों की आवृत्ति की गणना की जाती है, इससे कोई फर्क नहीं पड़ता कि कौन सा कॉलम `मान` पैरामीटर में पारित किया गया है, यह अंतिम परिणाम को प्रभावित नहीं करेगा:

```python
df.pivot_table(index = 'School',
               columns = 'Transfer',
               values = 'Name',
               aggfunc = 'count')
```

ऊपर से, हम देख सकते हैं कि इन दो कार्यों के बीच का अंतर यह है कि `क्रॉसस्टैब` की संबंधित स्थिति एक विशिष्ट अनुक्रम में गुजरती है, जबकि `पिवोट_टेबल` यदि अनुक्रम के अनुरूप मूल्य है तो कॉल की गई तालिका के नाम से गुजरती है में पारित, एक त्रुटि की सूचना दी जाएगी।

डिफ़ॉल्ट `गिनती` आँकड़ों के अलावा, सभी कस्टम फ़ंक्शन जो स्ट्रिंग और रिटर्न स्केलर को एकत्रित करते हैं, उपलब्ध हैं, जैसे संबंधित संयोजन की औसत ऊंचाई की गणना करना:

```python
pd.crosstab(index = df.School, columns = df.Transfer, values = df.Height, aggfunc = 'mean')
```
